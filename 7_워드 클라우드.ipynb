```python
import matplotlib.font_manager as fm

sys_font = fm.findSystemFonts()
[f for f in sys_font if 'Nanum' in f]
```

```python
from konlpy.corpus import kolaw
```

```python
text = kolaw.open('constitution.txt').read()
```

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
noun_text = okt.nouns(text)
print(noun_text)
```

```python
from collections import Counter
```

```python
count = Counter(noun_text)
print(count)
print(type(count))
```

```python
data = count.most_common(500) # 상위 500개, 튜플 구조로 반환
data = dict(data) # 튜플을 딕셔너리로 변환
print(data)
```

```python
# 딕셔너리에 저장된 데이터는 generate_from_frequencies() 메소드를 사용하여 워드 클라우드를 생성
wordcloud = WordCloud(font_path = '/usr/share/fonts/truetype/nanum/NanumPen.ttf',
                      background_color='ivory').generate_from_frequencies(data)

plt.figure(figsize = (7, 5))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

- 위 문제를 한나눔을 사용해서 풀기!

```python
from konlpy.tag import Hannanum
```

```python
hannanum = Hannanum()
```

```python
text = kolaw.open('constitution.txt').read()
```

```python
stop_words = ['제', '월', '일', '조', '수', '때', '그', '이', '바', '및', '안', '정', '관', '위', '장']
```

```python
# stopwords 옵션이 한글은 동작이 되지 않는다.
noun_text = hannanum.nouns(text)
noun_text = [each_word for each_word in noun_text if each_word not in stop_words]
print(noun_text)
```

```python
from collections import Counter
```

```python
count = Counter(noun_text)
print(count)
```

```python
data = count.most_common(500)
data = dict(data)
print(data)
```

```python
mask = np.array(Image.open('/content/drive/MyDrive/파이썬AI프로젝트202307/3_데이터 분석/korea_mask.jpg'))
```

```python
wordcloud = WordCloud(font_path = '/usr/share/fonts/truetype/nanum/NanumPen.ttf', background_color = 'ivory',
                      mask = mask).generate_from_frequencies(data)

plt.figure(figsize = (10, 15))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
