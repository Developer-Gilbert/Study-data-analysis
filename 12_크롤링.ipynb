- 크롤링(Crawling) : 인터넷의 데이터를 활용하기 위해 정보들을 분석하고 활용할 수 있게 수집하는 행위
- 스크래핑(Scraping) : 크롤링 + 데이터를 추출해서 가공하는 최종 목표

# 1. basicenglishspeaking.com

```python
import requests
from bs4 import BeautifulSoup
```

```python
site = 'https://basicenglishspeaking.com/daily-english-conversation-topics/'
request = requests.get(site)
print(request) # 200 이 정상 접속
print(type(request.text))
```

```python
soup = BeautifulSoup(request.text)
```

```python
type(soup)
```

```python
divs = soup.findAll('div', {'class':'thrv-columns'})
print(divs)
```

```python
# findAll 의 결과는 리스트 타입으로 가지고 오기 때문에
# 우리가 바로 그 리스트에서 findAll 은 실행할 수 없다.
for div in divs:
  links = div.findAll('a')
  print(links)
```

```python
links = divs[0].findAll('a')
links[0].text
```

### 간단한 문제!
- 출력예시와 똑같게 출력되도록 코드를 작성하시오.

  - 총 00 개의 주제를 찾았습니다.
  - 출력 예시     
     1, Family            
     2, Restaurant            
    ...                
    10, Book


```python
# 반복문을 이용해서 텍스트만 추출한 후에,
# 하나씩 topics 에 append 한다.
topics = []

links = divs[0].findAll('a')
for link in links:
  topics.append(link.text)

# 반복문을 활용하여 topics 에 있는 요소를 형식에 맞게 출력한다.
print('총', len(topics), '개의 주제를 찾았습니다.')
for i in range(len(topics)):
  print('{:2d}, {}'.format((i+1), topics[i]))
```

# 2. 다음 뉴스기사

```python
# https://v.daum.net/v/20231010164052131
# https://v.daum.net/v/20231010162810555

# 다음 뉴스 기사의 제목을 리턴해주는 함수
def daum_news_title(news_id):
  url = 'https://v.daum.net/v/{}'.format(news_id)
  request = requests.get(url)
  soup = BeautifulSoup(request.text)
  title = soup.find('h3', {'class':'tit_view'})
  if title:
    return title.text.strip()
  return '제목없음'
```

```python
print(daum_news_title(20231010164052131))
```

```python
print(daum_news_title(20231010162810555))
```

```python
print(daum_news_title(2023101016281))
```

# 3. 벅스뮤직 차트

```python
request = requests.get('https://music.bugs.co.kr/chart')
soup = BeautifulSoup(request.text)
```

```python
titles = soup.findAll('p', {'class' : 'title'})
# print(titles)
artists = soup.findAll('p', {'class' : 'artist'})
# print(artists)
```

```python
# 길이가 똑같은 것을 확인!
# 반복문 하나만 사용해서 원하는 데이터를 출력하자
print(len(titles))
print(len(artists))
```

```python
# 불필요한 공백과 줄바꿈이 있는 걸 확인~!
# 아래 코드셀에서 고쳐보자!
for i in range(len(titles)):
  title = titles[i].text
  artist = artists[i].text
  print(f'{i+1}위 {title} - {artist}')
```

```python
# strip() 으로 공백을 제거하니, 줄바꿈도 제거 됨
titles[1].text.strip()

# strip() 으로 공백을 제거하니, 줄바꿈도 제거 됨
artists[0].text.strip()

# 가수가 2명일 때, 이런 식으로 해결했다.
artists[57].text.strip().split('\n')[0]
```

```python
# 최종
for i in range(len(titles)):
  title = titles[i].text.strip()
  artist = artists[i].text.strip().split('\n')[0]
  print(f'{i+1}위 {title} - {artist}')
```

# 4. 멜론 차트

```python
# 무분별한 크롤링 막기 위해 멜론에서 장치를 해둔 것!
url = 'https://www.melon.com/chart/index.htm'
request = requests.get(url)
print(request)
```

```python
# Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36
header = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
request = requests.get(url, headers=header)
print(request)
```

```python
soup = BeautifulSoup(request.text)
```

```python
# 제목을 저장해둘 리스트
titles = soup.findAll('div', {'class', 'rank01'})
# print(titles)
# 가수를 저장해둘 리스트
artists = soup.findAll('span', {'class', 'checkEllipsis'})
print(artists)
```

```python
# 반복문을 활용하여 출력
# 최종
for i in range(len(titles)):
  title = titles[i].text.strip().replace('\n','')
  artist = artists[i].text.strip()
  print(f'{i+1:3d}위 {title} - {artist}')
```

```python
# 19금 삭제하기
for i in range(len(titles)):
  artist = artists[i].text.strip()
  title = titles[i].text.strip()

  if title.split('\n')[0] == '19금':
    title = title.split('\n')[1]
    print(f'{i+1:3d}위 {title} - {artist}')
  else:
    print(f'{i+1:3d}위 {title} - {artist}')
```
