# 문제
## folium 을 활용하여 컴퓨터 학원의 위치에 Marker를 표시해보자!
- folium.Map() 에 설정하는 위도, 경도는 데이터셋의 평균으로 설정
- folium.Pop 은 상호명과 도로명주소를 연결하여 설정
  - 예) 코리아IT아카데미-서울특별시 테헤란로 XXX
- for 문을 활용해주세요!

```python
data = shop_seoul_edu_computer

edu_map = folium.Map(location = [data.위도.mean(), data.경도.mean()], zoom_start = 12)

for i in data.index:
  # 팝업 설정
  edu_name = data.loc[i, '상호명'] + '-' + data.loc[i, '도로명주소']
  popup = folium.Popup(data.loc[i, '상호명'] + '-' + data.loc[i, '도로명주소'], max_width=200)

  # Map 에 add
  folium.Marker(location=[data.loc[i, '위도'], data.loc[i, '경도']], popup=popup).add_to(edu_map)

edu_map
```
