```python
import folium
```

```python
# folium.Map(location = [위도, 경도], zoom_start = 배율, ...)
map_folium = folium.Map(location = [37.500065615475805, 127.03556106196824], zoom_start = 17)
# save() : 작성된 지도차트를 html로 저장
map_folium.save('./map.html')
map_folium
```

```python
# 지도 스타일 (Open Street Map (기본값), Stamen Terrain, Stamen Toner)
map_folium = folium.Map(location = [37.500065615475805, 127.03556106196824], zoom_start = 17, tiles = 'Stamen Terrain')
map_folium
```

```python
map_folium = folium.Map(location = [37.500065615475805, 127.03556106196824], zoom_start = 17, tiles = 'Stamen Toner')
map_folium
```

```python
# Marker() : location 속성으로 지정한 위치에 마커를 생성
# Popup() : 마커를 클릭했을 때 표시할 문자열을 설정
# add_to() : 작성된 마커를 지도에 추가
popup = folium.Popup('코리아 IT 아카데미', max_width = 200)
map_folium = folium.Map(location = [37.500065615475805, 127.03556106196824], zoom_start = 17)
folium.Marker(location = [37.500065615475805, 127.03556106196824], popup = popup).add_to(map_folium) # 맵에 이름이 있어야 추가 가능
map_folium
```

```python
popup = folium.Popup('류호근 강사 생가', max_width = 200)
map_folium = folium.Map(location = [37.500065615475805, 127.03556106196824], zoom_start = 17)
folium.Marker(location = [34.940512780836194, 128.1989267283716], popup = popup).add_to(map_folium) # 맵에 이름이 있어야 추가 가능
map_folium
```

```python
map_folium = folium.Map(location=[37.56652479270633, 126.977], zoom_start=15)

popup = folium.Popup('서울특별시청', max_width=200)
folium.Marker(location=[37.56652479270633, 126.9779129116262], popup=popup).add_to(map_folium)

popup = folium.Popup('덕수궁', max_width=200)
folium.Marker(location=[37.56585320879458, 126.97514572197248], popup=popup).add_to(map_folium)

popup = folium.Popup('서울지방경찰청', max_width=200)
folium.Marker(location=[37.57504045678939, 126.97195576938117], popup=popup).add_to(map_folium)

popup = folium.Popup('광화문', max_width=200)
folium.Marker(location=[37.57617987638928, 126.97694467807803], popup=popup).add_to(map_folium)

popup = folium.Popup('영풍문고-종로본점', max_width=200)
folium.Marker(location=[37.56999947440024, 126.98228391170936], popup=popup).add_to(map_folium)

map_folium
```

```python
# icon : Marker 안에 있는 매개변수의 이름
# folium.Icon() : 원하는 색과 아이콘을 선택
# https://fontawesome.com/icons

map_folium = folium.Map(location=[37.56652479270633, 126.977], zoom_start=15)

popup = folium.Popup('서울특별시청', max_width=200)
folium.Marker(location=[37.56652479270633, 126.9779129116262], popup=popup,
              icon = folium.Icon(color = 'red', icon = 'star')).add_to(map_folium)

popup = folium.Popup('덕수궁', max_width=200)
folium.Marker(location=[37.56585320879458, 126.97514572197248], popup=popup,
              icon = folium.Icon(color = 'red', icon = 'check')).add_to(map_folium)

popup = folium.Popup('서울지방경찰청', max_width=200)
folium.Marker(location=[37.57504045678939, 126.97195576938117], popup=popup,
              icon = folium.Icon(color = 'orange', icon = 'ok')).add_to(map_folium)

popup = folium.Popup('광화문', max_width=200)
folium.Marker(location=[37.57617987638928, 126.97694467807803], popup=popup,
              icon = folium.Icon(color = 'black', icon = 'user')).add_to(map_folium)

popup = folium.Popup('영풍문고-종로본점', max_width=200)
folium.Marker(location=[37.56999947440024, 126.98228391170936], popup=popup).add_to(map_folium)

map_folium
```

```python
# CircleMarker() : 원 모양의 마커를 생성

map_folium = folium.Map(location=[37.56652479270633, 126.977], zoom_start=15)

popup = folium.Popup('서울시', max_width = 200)
# radius : 원의 크기
# color : 테두리색
# fill_color : 채우는 색
folium.CircleMarker(location=[37.56652479270633, 126.977], popup = popup,
                    radius = 200, color = 'blue', fill_color = 'red').add_to(map_folium)

map_folium
```
