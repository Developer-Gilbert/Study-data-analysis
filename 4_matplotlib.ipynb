# **2. 스타일 옵션**

```python
plt.figure(figsize = (3, 4)) # inch
plt.plot([1, 2, 3], [1, 2, 3])
plt.plot([1, 2, 3], [2, 4, 6])

# label 붙이는 방법
plt.title('타이틀 테스트', fontsize = 30) # 픽셀
plt.xlabel('x축 테스트', fontsize = 10)
plt.ylabel('y축 테스트', fontsize = 10, rotation = 45) # 각도

plt.show()
```

```python
plt.figure(figsize = (15, 10))

# color 옵션은, 앞글자 하나만 작성해도 나오는 색이 있다.
# linestyle : 선의 스타일
# marker : 데이터가 찍히는 지점을 어떤 모양으로 표시할 지
# markersize : marker 의 크기
# alpha : 투명도
plt.title('마커 설정', fontsize = 30, loc = 'right')
plt.plot(np.arange(10), np.arange(10), color = 'y', linestyle = '', marker = 'o')
plt.plot(np.arange(10), np.arange(10)*2, color = 'red', linestyle = '--', marker = 'v', markersize = 15)
plt.plot(np.arange(10), np.arange(10)*3, color = 'b', linestyle = '-.', marker = '*', markersize = 20, alpha = 0.5)

# 범례 추가
plt.legend(['10', '10*2', '10*3'], fontsize = 15, loc = 'lower right', ncol = 3)

# x축과 y축의 범위를 설정
plt.xlim(0, 12)
plt.xticks(rotation = 45)
plt.ylim(0, 35)
plt.yticks(rotation = 45)

plt.xlabel('강아지 나이', fontsize = 30)
plt.ylabel('IQ', fontsize = 30, rotation = 45) # 각도

plt.grid()
plt.show()
```

```python
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Korean']
y = [40, 80, 80, 50, 20, 90]

plt.figure(figsize = (4, 3))
# plt.bar(x, y, color = 'red', alpha = 0.7, align = 'edge')
plt.bar(x, y, color = 'red', alpha = 0.7, align = 'center')
plt.xticks(fontsize = 4)

plt.show()
```

```python
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Korean']
y = [40, 80, 80, 50, 20, 90]

plt.figure(figsize = (4, 3))
# plt.bar(x, y, color = 'red', alpha = 0.7, align = 'edge')
plt.barh(x, y, color = '#B3EFE7', alpha = 0.7, align = 'center')

plt.show()
```
